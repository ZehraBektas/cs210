# EDA and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

steps_data_copy = stepsData.copy()

steps_data_copy.set_index('creationDate', inplace=True)

# Plotting the distribution of step counts over time
plt.figure(figsize=(12, 6))
plt.hist(steps_data_copy.index, bins='auto', color='skyblue', edgecolor='black')
plt.title('Distribution of Step Counts Over Time')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

merged_data_step_flight = pd.merge(stepsData, FlightsClimbed, on='creationDate', how='inner')

correlation_steps_flights = merged_data_step_flight['step_count'].corr(merged_data_step_flight['flights_climbed'])

print(f"Correlation between Step Count and Flights Climbed: {correlation_steps_flights}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='step_count', y='flights_climbed', data=merged_data_step_flight)
plt.title('Correlation between Step Count and Flights Climbed')
plt.xlabel('Daily Step Count')
plt.ylabel('Flights Climbed')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

merged_data_step_walking = pd.merge(stepsData, WalkingRunningData, on='creationDate', how='inner')

correlation_steps_distance = merged_data_step_walking['step_count'].corr(merged_data_step_walking['distance'])

print(f"Correlation between Step Count and Walking/Running Distance: {correlation_steps_distance}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='step_count', y='distance', data=merged_data_step_walking)
plt.title('Correlation between Step Count and Walking/Running Distance')
plt.xlabel('Daily Step Count')
plt.ylabel('Walking/Running Distance')
plt.show()
```

```python
import numpy as np

features = stepsData[['creationDate', 'step_count']]

features.set_index('creationDate', inplace=True)

features = features.copy()
features.loc[:, 'activity'] = np.where(features['step_count'] > 800, 'active', 'inactive')

total_days = len(features)
active_days = features[features['activity'] == 'active'].shape[0]
percentage_active = (active_days / total_days) * 100

print(f'Percentage of Active Days: {percentage_active}%')

# Displaying the labeled days
print(features)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x=features.index, y='step_count', hue='activity', data=features, palette={'active': 'green', 'inactive': 'red'})
plt.title('Step Counts Over Time with Activity Labeling')
plt.xlabel('Date')
plt.ylabel('Step Count')
plt.xticks(rotation=45)
plt.show()
```

```python
kcal_per_km = 100

WalkingRunningData['EnergyBurned'] = WalkingRunningData['distance'] * kcal_per_km
```

```python
plt.figure(figsize=(10, 6))
plt.plot(WalkingRunningData['creationDate'], WalkingRunningData['EnergyBurned'], marker='o', linestyle='-', color='b')

plt.title('Energy Burned Over Time')
plt.xlabel('Date')
plt.ylabel('Energy Burned (kcal)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

max_energy_date = WalkingRunningData.loc[WalkingRunningData['EnergyBurned'].idxmax(), 'creationDate']
max_energy_value = WalkingRunningData['EnergyBurned'].max()

print(f"The date with the most energy consumption is {max_energy_date} with {max_energy_value} kcal.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

decibel_limit = 85

AudioData['creationDate'] = pd.to_datetime(AudioData['creationDate'])

AudioData['exceedance'] = AudioData['voice_desibel'] > decibel_limit

percentage_exceedance = (AudioData['exceedance'].mean()) * 100

print(f"Percentage of time exceeding {decibel_limit} decibels: {percentage_exceedance:.2f}%")

plt.figure(figsize=(12, 6))
plt.plot(AudioData['creationDate'], AudioData['voice_desibel'], label='Voice Decibel')
plt.axhline(y=decibel_limit, color='r', linestyle='--', label=f'Decibel Limit ({decibel_limit} dB)')
plt.fill_between(AudioData['creationDate'], 0, 1, where=AudioData['exceedance'], color='red', alpha=0.3, label='Exceedance')
plt.xlabel('Creation Date')
plt.ylabel('Voice Decibel')
plt.title('Voice Decibel and Exceedance Over Time')
plt.legend()
plt.show()
```
